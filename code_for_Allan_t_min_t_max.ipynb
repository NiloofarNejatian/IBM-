```python
yearly_mean_temperatures = yearly_data.mean(axis=1, skipna=True)
```

```python
yearly_mean_temperatures
```

```python
from sklearn.linear_model import LinearRegression
X = np.array(pd.DatetimeIndex(yearly_mean_temperatures.index).year).reshape(-1, 1)
y = yearly_mean_temperatures.values

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the values
trend = model.predict(X)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(X, y, color='lightblue', label='Yearly Mean Temperature')
plt.plot(X, trend, color='red', label='Trend Line')

# Adding titles and labels
plt.title('Overall PRCP Trend from 1890 to 2017')
plt.xlabel('Year')
plt.ylabel('Mean PRCP Across All Stations')
plt.legend()
plt.show()

# The slope of the trend line indicates the direction of the temperature trend
slope = model.coef_[0]
print(f"The slope of the trend line is {slope:.4f}, "
      f"indicating {'an increase' if slope > 0 else 'a decrease'} in temperature over time.")
```

```python
plt.figure(figsize=(15, 7))

for column in yearly_data.columns:
    # Skip the date column if it's included in the DataFrame
    if column == 'combined_date':
        continue

    # Plot each column's data with a separate call to plt.plot()
    plt.plot(yearly_data.index, yearly_data[column], label=column)

# Set the title and labels
plt.title('Minimum Temperature of Each Station Over Time')
plt.xlabel('Time')
plt.ylabel('Min Temperature')

# Add a legend to distinguish different stations
plt.legend()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(15, 7))

# Iterate over each station/column in the DataFrame
for station in yearly_data.columns:
    # Make sure to skip any non-station columns
    if station.lower() == 'combined_date' or station.lower() == 'date':
        continue

    # Obtain the minimum temperature for each station and its corresponding date
    min_temp = yearly_data[station].min()
    min_temp_date = yearly_data[station].idxmin()

    # Now plot this as a scatter plot point
    plt.scatter(min_temp_date, min_temp, label=f'{station} (Min Temp: {min_temp})')

# Set the title and labels
plt.title('MIN PRCP Recorded at Each Station')
plt.xlabel('Year')
plt.ylabel('PRCP ')  # Replace with correct unit

# Add a legend outside the plot to the right
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot with a tight layout
plt.tight_layout()
plt.show()
```
